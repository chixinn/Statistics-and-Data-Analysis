```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
stats.t.ppf(0.975,df=1)
stats.t.ppf(0.975,df=10)
```

```python
np.random.seed(9309)
n=100 # 样本数量
samples=[np.random.normal(loc=0,scale=3,size=100)for _ in range(n)]
fig,ax=plt.subplots(figsize=(10,7))
for i in np.arange(1,n,1):
    sample_mean=np.mean(samples[i])#计算样本均值
    se=stats.sem(samples[i])#std err标准差
    h=se*stats.t.ppf((1+0.95)/2,len(samples[i])-1)#t的分位点函数
    sample_ci=[sample_mean-h,sample_mean+h]
    if(sample_ci[0]<=0)and(0<=sample_ci[1]):
        plt.plot((sample_ci[0],sample_ci[1]),(i,i),color='blue')
        plt.plot(np.mean(samples[i]),i,'bo')#画中间这个点
    else:
        plt.plot((sample_ci[0],sample_ci[1]),(i,i),color='red')
        plt.plot(np.mean(samples[i]),i,'ro')
plt.axvline(x=0,ymin=0,ymax=1)#中间这条竖线
```
