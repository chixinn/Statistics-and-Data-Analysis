```python
def Ez1(p,q,n1):
    return (n1*p)/(p+q-p*q)
def Ez2(p,q,n2):
    return (n2*p)/(1-q+p*q)
```

```python
n=1000;n1=432;n2=232;n3=336
p0=0.5;q0=0.5
N=1000#max iteration times
emisilon=0.00001#error bound
i=0
while True:
    print(i,p0,q0)
    ptmp=p0
    qtmp=q0
    A=Ez1(p0,q0,n1)+Ez2(p0,q0,n2)
    B=Ez1(p0,q0,n1)
    C=Ez2(p0,q0,n2)
    p0=(A)/n
    q0=(n+n1-n2-B)/(2*n-A)
    # print(i,p0,q0)
    i=i+1
    if ((abs(qtmp-q0)<emisilon) and (abs(ptmp-p0)<emisilon)) or(i>N):
        break
print(p0,q0)
```
